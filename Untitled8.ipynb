```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def parse_wrk_report(file_path):
    with open(file_path, 'r') as f:
        data = f.read()

    # Extract Requests/sec
    reqs_per_sec = re.search(r'Requests/sec:\s+([\d\.]+)', data)
    reqs_per_sec = float(reqs_per_sec.group(1)) if reqs_per_sec else None

    # Extract Transfer/sec (convert to MB/s if needed)
    transfer_per_sec = re.search(r'Transfer/sec:\s+([\d\.]+)([KMG]?B)', data)
    if transfer_per_sec:
        val = float(transfer_per_sec.group(1))
        unit = transfer_per_sec.group(2)
        factor = {'KB': 1/1024, 'MB': 1, 'GB': 1024}
        transfer_per_sec = val * factor.get(unit, 1)
    else:
        transfer_per_sec = None

    # Extract latency avg and max (in ms)
    latency_avg = re.search(r'Latency\s+([\d\.]+)(\w+)', data)
    latency_max = re.findall(r'Latency.*Max\s+([\d\.]+)(\w+)', data)

    # Alternate method since latency max might be hard to find reliably:
    # From thread stats section:
    thread_stats = re.findall(r'Latency\s+([\d\.]+)(\w+)\s+([\d\.]+)(\w+)\s+([\d\.]+)(\w+)', data)
    if thread_stats:
        # Use first matched line for avg and max latency
        avg_val, avg_unit, stdev_val, stdev_unit, max_val, max_unit = thread_stats[0]
        def to_ms(val, unit):
            unit = unit.lower()
            if unit == 'us':
                return float(val) / 1000
            elif unit == 'ms':
                return float(val)
            elif unit == 's':
                return float(val) * 1000
            else:
                return float(val)  # fallback

        latency_avg = to_ms(avg_val, avg_unit)
        latency_max = to_ms(max_val, max_unit)
    else:
        latency_avg = None
        latency_max = None

    return {
        'Requests/sec': reqs_per_sec,
        'Latency Avg (ms)': latency_avg,
        'Latency Max (ms)': latency_max,
        'Transfer MB/sec': transfer_per_sec
    }


# Parse both reports
blocking_report = parse_wrk_report('block1kb.txt')
nonblocking_report = parse_wrk_report('nonblock1kb.txt')

# Combine into DataFrame for visualization
df = pd.DataFrame([
    {'Test Type': 'Blocking I/O', **blocking_report},
    {'Test Type': 'Non-blocking I/O', **nonblocking_report}
])

print(df)

# Plotting using seaborn
sns.set(style="whitegrid")

fig, axes = plt.subplots(2, 2, figsize=(12,8))

# Requests/sec
sns.barplot(x='Test Type', y='Requests/sec', data=df, ax=axes[0,0], palette='muted')
axes[0,0].set_title('Requests per Second')

# Latency Avg
sns.barplot(x='Test Type', y='Latency Avg (ms)', data=df, ax=axes[0,1], palette='muted')
axes[0,1].set_title('Average Latency (ms)')

# Latency Max
sns.barplot(x='Test Type', y='Latency Max (ms)', data=df, ax=axes[1,0], palette='muted')
axes[1,0].set_title('Max Latency (ms)')

# Transfer MB/sec
sns.barplot(x='Test Type', y='Transfer MB/sec', data=df, ax=axes[1,1], palette='muted')
axes[1,1].set_title('Transfer MB/sec')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def load_dstat_file(filename):
    df = pd.read_csv(filename, skiprows=1)
    # Convert time if needed, but here we ignore time axis in plotting
    return df

blocking_df = load_dstat_file('block1kbop.csv')
nonblocking_df = load_dstat_file('nonblock1kbop.csv')

# Metrics to compare
metrics = ['usr', 'sys', 'idl', 'writ', 'int', 'csw']  # add/write exact column names you want

# Compute mean values for each metric
blocking_means = blocking_df[metrics].mean()
nonblocking_means = nonblocking_df[metrics].mean()

# Prepare bar chart
labels = metrics
x = np.arange(len(labels))  # label locations
width = 0.35  # bar width

fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(x - width/2, blocking_means, width, label='Blocking I/O')
rects2 = ax.bar(x + width/2, nonblocking_means, width, label='Non-blocking I/O')

ax.set_ylabel('Average Value')
ax.set_title('Average System Metrics: Blocking vs Non-blocking I/O')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
ax.grid(True, axis='y')

# Label bars with values
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0,3), textcoords='offset points',
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re
import numpy as np
import seaborn as sns

# --- Utility: Parse wrk txt output ---
def parse_wrk_file(filename):
    with open(filename, 'r') as f:
        data = f.read()

    req_sec_match = re.search(r"Requests/sec:\s+([\d.]+)", data)
    req_sec = float(req_sec_match.group(1)) if req_sec_match else np.nan

    transfer_match = re.search(r"Transfer/sec:\s+([\d.]+)([KMG]?B)", data)
    if transfer_match:
        val = float(transfer_match.group(1))
        unit = transfer_match.group(2)
        if unit == 'KB':
            transfer = val / 1024
        elif unit == 'MB':
            transfer = val
        elif unit == 'GB':
            transfer = val * 1024
        else:
            transfer = val
    else:
        transfer = np.nan

    latency_match = re.search(r"Latency\s+([\d.]+)([mun]?s)", data)
    if latency_match:
        val = float(latency_match.group(1))
        unit = latency_match.group(2)
        if unit == 'ms':
            latency = val
        elif unit == 'us':
            latency = val / 1000
        elif unit == 'ns':
            latency = val / 1e6
        else:
            latency = val
    else:
        latency = np.nan

    return {'Requests/sec': req_sec, 'Transfer(MB/s)': transfer, 'Avg Latency(ms)': latency}

# --- Load dstat csv file ---
def load_dstat_csv(filename):
    df = pd.read_csv(filename, skiprows=1)
    # Select meaningful metrics (adjust column names if needed)
    cols = ['usr', 'sys', 'idl', 'writ', 'int', 'csw']
    cols = [c for c in cols if c in df.columns]
    avg_metrics = df[cols].mean()
    return avg_metrics

# --- Function to run analysis and plotting ---
def analyze_and_plot(wrk_files, dstat_files, labels=None):
    if labels is None:
        labels = [f"Test {i+1}" for i in range(len(wrk_files))]

    wrk_results = {}
    dstat_results = {}

    for label, wrk_file, dstat_file in zip(labels, wrk_files, dstat_files):
        wrk_results[label] = parse_wrk_file(wrk_file)
        dstat_results[label] = load_dstat_csv(dstat_file)

    wrk_df = pd.DataFrame(wrk_results).T
    dstat_df = pd.DataFrame(dstat_results).T
    combined_df = pd.concat([wrk_df, dstat_df], axis=1)

    print("=== Combined Summary ===")
    print(combined_df)

    # wrk Requests/sec & Latency bar chart
    fig, ax1 = plt.subplots(figsize=(12, 6))
    ax1.bar(combined_df.index, combined_df['Requests/sec'], alpha=0.7, label='Requests/sec')
    ax1.set_ylabel('Requests/sec')
    ax1.set_ylim(0, combined_df['Requests/sec'].max() * 1.2)

    ax2 = ax1.twinx()
    ax2.plot(combined_df.index, combined_df['Avg Latency(ms)'], 'ro-', label='Avg Latency (ms)')
    ax2.set_ylabel('Latency (ms)')
    ax2.set_ylim(0, combined_df['Avg Latency(ms)'].max() * 1.5)

    ax1.set_title('wrk Test Results Comparison')
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    plt.grid(True)
    plt.show()

    # CPU & Context Switch comparison bar chart
    metrics_to_plot = ['usr', 'sys', 'idl', 'writ', 'int', 'csw']
    dstat_df[metrics_to_plot].plot(kind='bar', figsize=(14, 7), title="Average System Metrics (Dstat) Comparison")
    plt.ylabel("Average Value")
    plt.grid(axis='y')
    plt.show()

    # Correlation heatmap
    plt.figure(figsize=(10,8))
    sns.heatmap(combined_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Correlation Between wrk and System Metrics")
    plt.show()

# --- Example usage ---
# Provide your actual filenames here:
wrk_files = [
    'block1kb.txt',
    'nonblock1kb.txt',
    'block8kb.txt',
    'nonblock8kb.txt'
]

dstat_files = [
    'block1kbop.csv',
    'nonblock1kbop.csv',
    'block8kbop.csv',
    'nonblock8kbop.csv'
]

labels = [
    'Blocking 1KB',
    'Non-blocking 1KB',
    'Blocking 8KB',
    'Non-blocking 8KB'
]

# Run analysis
analyze_and_plot(wrk_files, dstat_files, labels)
```
